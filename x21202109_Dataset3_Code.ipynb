## Data Pre-Processing 

```python
#Converting datatypes to appropriate types
heartstroke_data["bmi"] = pd.to_numeric(heartstroke_data.bmi, errors='coerce')

heartstroke_data.astype({"gender":'category',"smoking_status":'category', "Residence_type":'category' , "work_type":'category' , "ever_married":'category'})
```

```python
#Identifying the missing values 

heartstroke_data.describe()
```

```python
#identifying the missing values with mean of bmi

def replace_with_mean(heartstroke_data):
    total_null = heartstroke_data.isnull().sum().sort_values(ascending=False)
    percent= (heartstroke_data.isnull().sum()/heartstroke_data.isnull().count()).sort_values(ascending=False)
    bmi_missing_data = pd.concat([total_null, percent], axis=1, keys=['Total', 'Percent'])
    return bmi_missing_data

replace_with_mean(heartstroke_data)
```

```python
# Imputing the missing values with the mean
heartstroke_data=heartstroke_data.fillna(np.mean(heartstroke_data['bmi']))
```

```python
#Visualising th effect of factor variables on stroke

data_cat=stroke_data[['gender', 'hypertension', 'heart_disease', 'ever_married','Residence_type']]

fig=plt.figure(figsize=(20,23))
background_color = '#f6f5f7'
fig.patch.set_facecolor(background_color) 
for indx,val in enumerate(data_cat.columns):
    ax=plt.subplot(4,2,indx+1)
    ax.set_facecolor(background_color)
    ax.set_title(val,fontweight='bold',fontfamily='serif')
    for i in ['top','right']:
        ax.spines[i].set_visible(False)
    ax.grid(linestyle=':',axis='y')
    sns.countplot(data_cat[val],palette='OrRd_r',hue=stroke_data['stroke'])
```
